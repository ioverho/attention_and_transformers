```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
word_vectors = [
    [ 2.0,  1.0],
    [-0.5,  1.5],
    [-1.0, -1.5]
]

word_vectors = np.array(word_vectors)
origin = np.zeros_like(word_vectors)

fig, ax = plt.subplots(1, 1, figsize=(5,5))

ax.quiver(
    origin[:, 0],
    origin[:, 1],
    word_vectors[:, 0],
    word_vectors[:, 1],
    angles="xy",
    pivot="tail",
    scale_units='xy',
    scale=1,
    color=["red", "green", "blue"]
)

ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)

# Move left y-axis and bottom x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.set_yticks([])
ax.set_xticks([])

ax.text(
    word_vectors[0, 0],
    word_vectors[0, 1],
    s="I",
    fontsize=16,
)

ax.text(
    word_vectors[1, 0],
    word_vectors[1, 1],
    s="am",
    fontsize=16,
)

ax.text(
    word_vectors[2, 0],
    word_vectors[2, 1]-0.15,
    s="Sam",
    fontsize=16,
)

fig.tight_layout()

fig.savefig(
    "../figures/word_vectors.svg",
    pad_inches=0.0,
    bbox_inches=0,
    transparent=True,
    )
```

```python
weights = [
    [0.6, 0.1, 0.3],
    [0.3, 0.5, 0.2],
    [0.2, 0.1, 0.7],
]

word_vectors = np.array(word_vectors)
weights = np.array(weights)

origin = np.zeros_like(word_vectors)

fig, ax = plt.subplots(1, 1, figsize=(5, 5))

ax.quiver(
    origin[:, 0],
    origin[:, 1],
    word_vectors[:, 0],
    word_vectors[:, 1],
    angles="xy",
    pivot="tail",
    scale_units='xy',
    scale=1,
    color=["lightgray", "lightgray", "lightgray"]
)

ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)

# Move left y-axis and bottom x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.set_yticks([])
ax.set_xticks([])

ax.text(
    word_vectors[0, 0],
    word_vectors[0, 1],
    s="I",
    c="gray",
    fontsize=16,
)

ax.text(
    word_vectors[1, 0],
    word_vectors[1, 1],
    s="am",
    c="gray",
    fontsize=16,
)

ax.text(
    word_vectors[2, 0],
    word_vectors[2, 1]-0.15,
    s="Sam",
    c="gray",
    fontsize=16,
)

final_positions = []
colours = ["red", "green", "blue"]
for i in range(len(colours)):
    position = np.zeros(2)

    for ii in range(len(colours)):
        dx = weights[i, (i + ii) % len(colours)] * word_vectors[(i + ii) % len(colours), 0]
        dy = weights[i, (i + ii) % len(colours)] * word_vectors[(i + ii) % len(colours), 1]

        ax.quiver(
            position[0],
            position[1],
            dx,
            dy,
            angles="xy",
            pivot="tail",
            scale_units='xy',
            scale=1,
            fc=["none"],
            ec=colours[(i + ii) % len(colours)],
            ls=["--"],
            linewidth=1,
        )

        position[0] = position[0] + dx
        position[1] = position[1] + dy

    final_positions.append(position)

    ax.quiver(
        final_positions[i][0],
        final_positions[i][1],
        angles="xy",
        pivot="tail",
        scale_units='xy',
        scale=1,
        fc=[colours[i]],
        ec=[colours[i]],
    )

ax.text(
    final_positions[0][0] + 0.1,
    final_positions[0][1] - 0.1,
    s="I$^\\prime$",
    c="black",
    fontsize=16,
)

ax.text(
    final_positions[1][0] + 0.1,
    final_positions[1][1] - 0.1,
    s="am$^\\prime$",
    c="black",
    fontsize=16,
)

ax.text(
    final_positions[2][0] + 0.1,
    final_positions[2][1] - 0.1,
    s="Sam$^\\prime$",
    c="black",
    fontsize=16,
)

fig.tight_layout()

fig.savefig(
    "../figures/adjusted_word_vectors.svg",
    pad_inches=0.0,
    bbox_inches=0,
    transparent=True,
    )

```
