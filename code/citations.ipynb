```python
# library
import matplotlib.pyplot as plt

fontsize = 20

# Create bars
barWidth = 0.9
heights = [1245, 5163, 10597, 18279, 28036, 41584, 54890, 9072]
x_vals = range(len(heights))
x_labels = [2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]

fig, ax = plt.subplots(1, 1, figsize=(1165 / 96, 189 / 89))

# Create barplot
ax.bar(x_vals, heights, width=barWidth, color="lightgray", edgecolor="black", linewidth=2)

# Years
ax.set_xticks(x_vals, x_labels, rotation=0, fontsize=fontsize)

max_y = max(heights)

# Text on the top of each bar
for x, y in zip(x_vals, heights):
    ax.text(x=x , y=y+0.05 * max_y, s=y, size=fontsize, horizontalalignment="center")

ax.set_yticks([])

ax.spines[["left", "right", "top"]].set_visible(False)
ax.spines[["bottom"]].set_linewidth(2)

# Adjust the margins
fig.tight_layout()

fig.savefig(
    "../figures/vaswani_et_al_citations_rate.svg",
    pad_inches=0.0,
    bbox_inches=0,
    transparent=True,
    )
```
